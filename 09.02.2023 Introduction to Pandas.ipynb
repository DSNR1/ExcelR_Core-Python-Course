### Visualization

```python
# Matplotlib
```

```python
from matplotlib import pyplot as plt
```

```python
cars_data.corr()
```

```python
plt.scatter(cars_data["Age"], cars_data["Price"], c = "red")
plt.title("Scatter plot of Price vs Age")
plt.xlabel("Age in months")
plt.ylabel("Price")
plt.show()
```

```python
plt.hist(cars_data["KM"], color= "green", bins= 10, edgecolor="white" )
plt.title("Histogram")
plt.xlabel("KM")
plt.ylabel("Frequency")
plt.figure(figsize=(10,8))
plt.savefig("bar.jpeg")
plt.show()
```
